# Google Search Analysis with Python

Google doesn’t give much access to the data about daily search queries, but another application of google known as Google Trends can be used for the task of Google search analysis. Google Trends provides an API that can be used to analyze the daily searches on Google. This API is known as pytrends, you can easily install it in your systems by using the pip command; pip install pytrends.

ES:Importamos librerias, prestamos atención a la libreria "pytrends.request"
EN:We import the libraries paying attention to the "pytrends.request" one.

```python
import pandas as pd
from pytrends.request import TrendReq  #API para analyzar diariamente las busquedas de google
import matplotlib.pyplot as plt
trends = TrendReq()
```

EN:Here I will be analyzing the Google search trends on the queries based on “Machine Learning”, so let’s create a DataFrame of the top 10 countries which search for “Deportivo la coruña” on Google.
ES: Analizamos las busquedas en google de "Deportivo la coruña", creamos un dataframe de los 10 paises con mas busquedas.

```python
trends.build_payload(kw_list=["Deportivo la coruña"]) #Ponemos como palabrra clave "Machine learning" en trends
data = trends.interest_by_region() # cargamos el dataframe 
data = data.sort_values(by="Deportivo la coruña", ascending=False) #Ordenamos valores de manera descendente.
data = data.head(10)
print(data)
```

EN:We can also visualize this data using a bar chart:
ES:Lo podemos analizar con diagrma de barras

```python
data.reset_index().plot(x="geoName", 
                        y="Deportivo la coruña", 
                        figsize=(15,12), kind="bar")
plt.style.use('fivethirtyeight')
plt.show()
```

EN:So as we all know that Machine Learning has been the focus of so many companies and students for the last 3-4 years, so let’s have a look at the trend of searches to see how the total search queries based on “Machine Learning” increased or decreased on Google
ES:Observemos como el trend "Machine learning" evoluciona a lo largo de los años.

```python
data = TrendReq(hl='en-US', tz=360)
data.build_payload(kw_list=['Deportivo la coruña'])
data = data.interest_over_time()
fig, ax = plt.subplots(figsize=(15, 12))
data['Deportivo la coruña'].plot()
plt.style.use('fivethirtyeight')
plt.title('Búsquedas Totales en Google de Deportivo la coruña', 
          fontweight='bold')
plt.xlabel('Año')
plt.ylabel('Cuenta total')
plt.show()
```
